```python
# Import necessary components for text summarization
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load a pretrained model and tokenizer designed for summarization
checkpoint = "sshleifer/distilbart-cnn-12-6"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)

# Display some stats related to the tokenizer
print("Maximum token length for the model:", tokenizer.model_max_length)
print("Maximum token length for a single sentence:", tokenizer.max_len_single_sentence)
print("Number of special tokens to add:", tokenizer.num_special_tokens_to_add())

# Tokenize the content into sentences
import nltk
nltk.download('punkt')  # Download the NLTK sentence tokenizer
sentences = nltk.tokenize.sent_tokenize(text)  # Tokenize the content into sentences

# Display the maximum token length of the longest sentence
print("Maximum token length of the longest sentence:", max([len(tokenizer.tokenize(sentence)) for sentence in sentences]))

# Split the content into chunks to fit the model's maximum token limit
length = 0
chunk = ""
chunks = []
count = -1
for sentence in sentences:
    count += 1
    combined_length = len(tokenizer.tokenize(sentence)) + length
    if combined_length <= tokenizer.max_len_single_sentence:
        chunk += sentence + " "
        length = combined_length
        if count == len(sentences) - 1:
            chunks.append(chunk.strip())
    else:
        chunks.append(chunk.strip())
        length = 0
        chunk = ""
        chunk += sentence + " "
        length = len(tokenizer.tokenize(sentence))

print("Number of chunks:", len(chunks))

# Display more stats about the chunks
print("Token lengths for each chunk:", [len(tokenizer.tokenize(c)) for c in chunks])
print("Input ID lengths for each chunk:", [len(tokenizer(c).input_ids) for c in chunks])

print("Total input ID length for all chunks:", sum([len(tokenizer(c).input_ids) for c in chunks]))
print("Input ID length for the entire text:", len(tokenizer(text).input_ids))
print("Total token length for all chunks:", sum([len(tokenizer.tokenize(c)) for c in chunks]))
print("Token length for the entire text:", len(tokenizer.tokenize(text)))

# Convert chunks into inputs for the model
inputs = [tokenizer(chunk, return_tensors="pt") for chunk in chunks]


```

#### Result

```python
# Use the model to generate summaries for each chunk and display them
# Here is where we make the change, using max_new_tokens instead of max_length
summary = []
for input in inputs:
    output_summary = model.generate(**input, max_new_tokens=tokenizer.model_max_length)
    summary.append(tokenizer.decode(*output_summary, skip_special_tokens=True))
print('\n'.join(summary))
```

Create a basic recommendation system using Hugging Face's model to generate
innovative ideas based on the extracted elements.

```python
def chunks_creator(tokenizer,FileContent, max_len ) :


  # extract the sentences from the document
  sentences = nltk.tokenize.sent_tokenize(FileContent)

  print(f'Total numer IDs : {sum([len(tokenizer(c).input_ids) for c in sentences])}\n')
  print(f'Token max : {max([len(tokenizer.tokenize(sentence)) for sentence in sentences])}')

  # Create the chunks
  # initialize
  length = 0
  chunk = ""
  chunks = []
  count = -1
  for sentence in sentences:
    count += 1
    combined_length = len(tokenizer.tokenize(sentence)) + length # add the no. of sentence tokens to the length counter

    if combined_length  <= max_len : # if it doesn't exceed
      chunk += sentence + " " # add the sentence to the chunk
      length = combined_length # update the length counter

      # if it is the last sentence
      if count == len(sentences) - 1:
        chunks.append(chunk.strip()) # save the chunk

    else:
      chunks.append(chunk.strip()) # save the chunk

      # reset
      length = 0
      chunk = ""

      # take care of the overflow sentence
      chunk += sentence + " "
      length = len(tokenizer.tokenize(sentence))
  print(f'len(chunks) : {len(chunks)}')

  return chunks
```

```python
from transformers import RobertaTokenizer, RobertaForQuestionAnswering
checkpoint_qa = 'distilbert-base-cased-distilled-squad'
# Cargar el modelo y el tokenizador
tokenizer_qa = RobertaTokenizer.from_pretrained("deepset/roberta-base-squad2")
model_qa = RobertaForQuestionAnswering.from_pretrained("deepset/roberta-base-squad2")
```

```python
chunks_qa = chunks_creator( tokenizer_qa , '\n'.join(paragraphs), 460 )
```

```python
len(chunks_qa)
```

```python
[len(tokenizer_qa.tokenize(c)) for c in chunks_qa ]
```

```python
def question_answer(contexts,question):

  print(f'Question : {question}')

  for context in contexts:
      inputs = tokenizer_qa.encode_plus(question, context, return_tensors="pt", truncation=True, max_length=512)
      outputs = model_qa(**inputs)
      answer_start_scores = outputs.start_logits
      answer_end_scores = outputs.end_logits
      answer_start = answer_start_scores.argmax()
      answer_end = answer_end_scores.argmax() + 1
      answer = tokenizer_qa.convert_tokens_to_string(tokenizer_qa.convert_ids_to_tokens(inputs["input_ids"][0][answer_start:answer_end]))
      # print(f"Context: {context}\n")
      if len(answer) > 3 :
        print(f"\n Answer: {answer}\n")
```

#### Result

```python
question_answer(chunks_qa, 'Where born Murakami ?' )
```
